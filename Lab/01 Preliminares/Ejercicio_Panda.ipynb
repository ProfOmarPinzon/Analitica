# Ejercicio de Introducción a Pandas

Importamos la biblioteca pandas y matplotlib para visualización

```python
import pandas as pd
import matplotlib.pyplot as plt
```

# 1. Crear un DataFrame básico

```python
print("1. Creando un DataFrame básico:")
datos = {
    'Nombre': ['Ana', 'Luis', 'Carlos', 'Sofía', 'María'],
    'Edad': [23, 34, 45, 29, 31],
    'Ciudad': ['Bogotá', 'Medellín', 'Cali', 'Bogotá', 'Cali'],
    'Puntaje': [85, 90, 88, 76, 95]
}
df = pd.DataFrame(datos)
print(df)
```

# 2. Filtrar datos por una condición

```python
print("\n 2. Filtrar estudiantes con puntaje mayor a 85:")
filtrado = df[df['Puntaje'] > 85]
print(filtrado)
```

# 3. Añadir una nueva columna calculada

```python
print("\n3. Añadir una columna que clasifica según la edad:")
df['Categoría Edad'] = ['Joven' if edad < 30 else 'Adulto' for edad in df['Edad']]
print(df)
```

# 4. Agrupar los datos por 'Ciudad' y calcular la media de 'Puntaje'

```python
print("\n4. Calcular la media de puntaje por ciudad:")
promedio_por_ciudad = df.groupby('Ciudad')['Puntaje'].mean()
print(promedio_por_ciudad)
```

# 5. Visualización: Gráfico de caja y bigotes

```python
print("\n5. Generar un gráfico de caja y bigotes para las edades:")
plt.figure(figsize=(8, 6))
plt.boxplot(df['Edad'], patch_artist=True, boxprops=dict(facecolor="lightblue"))
plt.title("Distribución de Edades (Caja y Bigotes)")
plt.ylabel("Edad")
plt.grid(True)
plt.show()
```

# 6. Visualización: Histograma de puntajes

```python
print("\n6. Generar un histograma de puntajes:")
plt.figure(figsize=(8, 6))
plt.hist(df['Puntaje'], bins=5, color="skyblue", edgecolor="black")
plt.title("Distribución de Puntajes")
plt.xlabel("Puntaje")
plt.ylabel("Frecuencia")
plt.grid(axis='y')
plt.show()
```

# 7. Guardar el DataFrame como un archivo CSV

```python
print("\n7. Guardando el DataFrame en un archivo CSV llamado 'estudiantes.csv'.")
df.to_csv('estudiantes.csv', index=False)

print("\n¡Ejercicio completado! Revisar el archivo 'estudiantes.csv' generado.")
```
